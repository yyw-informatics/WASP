## Filtering Data

### Remove non-productive sequences

```r
dat_file = 'data_db-pass.tsv'
data <- readChangeoDb(dat_file) %>% filter(productive)
cat(paste("There are", nrow(data), "sequences (rows) in the data.\n"))
```

### Remove cells with multiple heavy chains

```r
multi_heavy <- table(filter(data, locus == "IGH")$cell_id)
multi_heavy_cells <- names(multi_heavy)[multi_heavy > 1]
data <- filter(data, !cell_id %in% multi_heavy_cells)
cat(paste("There are", nrow(data), "rows in the data after filtering out cells with multiple heavy chains.\n"))
```

### Remove cells without heavy chains

```r
# split cells by heavy and light chains
heavy_cells <- filter(data, locus == "IGH")$cell_id
light_cells <- filter(data, locus == "IGK" | locus == "IGL")$cell_id
no_heavy_cells <- light_cells[which(!light_cells %in% heavy_cells)]

data <- filter(data, !cell_id %in% no_heavy_cells)
cat(paste("There are", nrow(data), "rows in the data after filtering out cells without heavy chains."))
```

```r
glimpse(data)
```

## Identify potentially novel v gene alleles


[`findNovelAlleles`](https://tigger.readthedocs.io/en/stable/topics/findNovelAlleles/) from package `TIgGER` analyzes mutation patterns in sequences thought to align to each germline allele in order to determine which positions might be polymorphic.
Output data.frame is explained here under `Value` section.

### Heavy chain V genes

```r
hdat <- data %>% filter(locus == "IGH")
ighv <- readIgFasta('imgt_human_IGHV.fasta')
nvh <- findNovelAlleles(hdat, germline_db = ighv, nproc = 8) # find novel alleles
```

```r
selectNovel(nvh)
```

##### One example

```r
nvh %>% filter(note == "Novel allele found!") %>% filter(row_number() == 1) %>% glimpse()
```

###### Visualize the evidence for novel alleles

[`plotNovel`](https://tigger.readthedocs.io/en/stable/topics/plotNovel/) produces the following three figures, each represent a piece of evidence.
1.   Top **line graph**: For all sequences aligned to a germline allele, the *mutation frequency* at each position along the aligned sequence as a function of the sequence-wide *mutation count*. 
  *   Each line represent a position. 
  *   The red line represents the position that pass the novel allele test
  *   Positions that contain polymorphisms (rather than somatic hypermutations) will exhibit a high apparent mutation frequency for a range of sequence-wide mutation counts. 
  *   Mutational hot/cold-spots are expected to have y-intercept around zero.
  *   The minimum y-intercept threshold can be specified by `y_intercept`, by default it is 0.125. 

2.   Middle **bar graph**: the nucleotide usage at each of the polymorphic position as a function of sequence-wide mutation count.  
  * Polymorphic allele are expected to be prevalent at all mutation counts
  * The mutation count is expected to be equal to the number of polymorphisms in the novel sequence to be the most prevalent.

3.   Bottom **stacked bar graph**: To avoid cases where a clonal expansion might lead to a false positive, TIgGER examines the combinations of J gene junction length among sequences which perfectly match the proposed germline allele. 
  *  Clonally related sequences usually share the same V gene, J gene, and the junction length.
  *  Requiring the novel allele to be found in different combinations of J gene and junction length is a proxy for requiring it to be found in different clonal lineages. 
  * The maximum portion of sequences which can consist of a specific combination of J gene and junction length may be specified by `j_max`.

```r
p <- c()
for (i in 1:length(nrow(selectNovel(nvh)))){
  p[[i]] <- plotNovel(hdat, selectNovel(nvh)[i,])
}
cat(nrow(selectNovel(nvh))," figures generated")
```

## Genotyping, including novel V gene alleles

[`inferGenotype`](https://tigger.readthedocs.io/en/stable/topics/inferGenotype/) infers genotypes, by finding the mimimum number set of alleles that can explain the majority of each gene's calls. 

*   The most commom allele of each gene is included in the genotype first, and the next most commom allele is added until the desired fraction of alleles can be explained. 

In addition, [`inferGenotypeBayesian`](https://tigger.readthedocs.io/en/stable/topics/inferGenotypeBayesian/) infers an subject’s genotype by applying a Bayesian framework with a Dirichlet prior for the multinomial distribution.


#### Heavy chain V genes

```r
gth <- inferGenotype(hdat, germline_db = ighv, novel = nvh)

# save genotype inf .fasta format to be used later with CreateGermlines.py
gthseq <- genotypeFasta(gth, ighv, nvh)
writeFasta(gthseq, "tigger_heavy_v_genotype.fasta")

gth %>% arrange(desc(total)) 
```

##### Visualize genotypes
[`plotGenotype`](https://tigger.readthedocs.io/en/stable/topics/plotGenotype/) Each row is a gene, with colored cells indicating each of the alleles for that gene that are included in the inferred genotype.

```r
plotGenotype(gth)
```

### Light chain lamda

```r
gtl <- inferGenotype(ldat, germline_db = iglv, novel = nvl)
gtlseq <- genotypeFasta(gtl, iglv, nvl)
writeFasta(gtlseq, "tigger_light_l_v_genotype.fasta")
gtl %>% arrange(desc(total)) 
```

##### Visualize genotypes

```r
plotGenotype(gtl)
```

### Light chain Kappa

```r
gtk <- inferGenotype(kdat, germline_db = igkv, novel = nvk)
gtkseq <- genotypeFasta(gtk, igkv, nvk)
writeFasta(gtkseq, "tigger_light_k_v_genotype.fasta")
gtk %>% arrange(desc(total)) 
```

##### Visualize genotypes

```r
plotGenotype(gtk)
```

## Reassign V gene allele annotations¶


[`reassignAlleles`](https://tigger.readthedocs.io/en/stable/topics/reassignAlleles/) corrects the V gene assignments (`v_call`) and produce a new column called `v_call_genotyped`.




```r
hdat_2 <- reassignAlleles(hdat, gthseq)
```

### Total corrected and examples

```r
corrected_v_calls <- hdat_2 %>% filter(v_call != v_call_genotyped) %>% select(v_call, v_call_genotyped) 
cat('A total of', nrow(corrected_v_calls), 'v calls are corrected, here are a few examples:' )
corrected_v_calls %>% sample_n(5)
```

### A before and after comparison (fractions)

```r
data.frame(Ambiguous=c(mean(grepl(",", hdat_2$v_call)),
                       mean(grepl(",", hdat_2$v_call_genotyped))),
           NotInGenotype=c(mean(hdat_2$v_call %in% not_in_genotype),
                           mean(hdat_2$v_call_genotyped %in% not_in_genotype)),
           row.names=c("Before", "After")) %>% 
    t() %>% round(3)
```
